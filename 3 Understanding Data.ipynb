## Conditionally Filtering Dataframes

```python
"""
You want to subset the age and al columns where age is 60
loc label-location based indexer  indexer for selection by label. 
You cannot use 
df[df['age'] == 60, ['age','al']]
"""

df.loc[df['Age'] == 60,['MonthlyRate', 'BusinessTravel']].head()

# conditionally filters the dataframe 
# that only includes rows whose age column is 60
# Then access the only two columns: bp and al
# Basically, this gives the of bp and al values of records whose age columns is 60
# In other words, this gives the bp and al values of 60 year old ppl

```

```python
"""
You can use multiple conditions with "&" sign. 
You want to subset the age and al columns where age is between 60 and 70
You can do the same thing in numpy arrays.
You need to know the index numbers as you cannot use column names 
df_as_array[((df_as_array[:,0]>60) & (df_as_array[:,0]<70))][:, [0,3]]
"""

df.loc[((df['Age'] > 40) & (df['Age'] < 50)), ['MonthlyRate', 'BusinessTravel']].head()


# conditionally filters the dataframe 
# that only includes rows whose age column is between 60 and 70
# Then accesses the only two columns: bp and al
# Basically, this gives the of bp and al values of records whose age columns is between 60 & 70
# In other words, this gives the bp and al values of ppl between 60 & 70
```

## Pandas Pivot Tables

```python
# imports the pandas and numpy libraries so we can work 
# w/ dataframes and arrays

import pandas as pd 
import numpy as np

# reads the movies dataset. 
# This dataset contains 1000 movies and info about them
movies = pd.read_csv('movies.csv') 
```

```python
# let's have a sneak peak at the dataset 
# (remember that head is a method that displays 
# n numbers of rows in a dataset as default in this case n = 2)
movies.head(2) 
```

```python
# fill in the missing values of the 'Revenue (Millions)' variable 
# using median of that variable
# remember median is the 50th percentile

movies['Revenue (Millions)'].fillna(movies['Revenue (Millions)'].median(), 
                                    inplace = True)


# groups by the dataset using the metascore variable
# selects the 'Revenue (Millions)' variable and 
# computes the mean of 'Revenue (Millions)' by metascore variable
# sort the results in descending order

movies.groupby('Metascore')['Revenue (Millions)'].mean().sort_values(
    ascending = False)
```

```python
# Creates a pivot table using the movies data
# where row indeces are year and metascore
# basically movies are grouped by using year and metascore
# then revenue is summed using with the aggfunc

pd.pivot_table(movies, values='Revenue (Millions)', 
               index=['Year', 'Metascore'], 
               aggfunc = np.sum)
```

```python

pd.pivot_table(movies, values='Revenue (Millions)', index=['Year'], aggfunc=np.sum) 
# gives teh total revenue per year
# this may be misleading as in some years there may be more movies. Average is a better statistic to use. 
```

```python
pd.pivot_table(movies, values='Revenue (Millions)', index=['Year'], aggfunc = 'count')
# count the number of movies in each year
```

```python
pivot_rev_by_year = pd.pivot_table(movies, values='Revenue (Millions)', index=['Year'], aggfunc=np.mean)
pivot_rev_by_year
# gives the average revenue per year
# looks like 2012 was a great year for Hollywood. 
```

```python
pivot_rev_by_year[pivot_rev_by_year.index == 2006]
```

## Aggregating and Summarizing Data

```python
movies.groupby('Year').count() # count how many movies out there per year
```

```python
movies.groupby('Year')['Revenue (Millions)'].mean() # group by year and get average revenue per year
```

```python
movies.groupby('Genre')['Revenue (Millions)'].mean() # group by genre and get the average revenue.
```

```python
# WOOOW - there are many genres for each movie. This is difficult to read. 
# Let's split this column and grab the first genra as the primary genra.


prim_genre = []
for i in movies['Genre'].str.split(','):
    prim_genre.append(i[0])

```

```python
prim_genre # display the primary genre list
```

```python
movies['primary_genre'] = prim_genre # create a new column and save the parimary genre into the dataframe
```

```python
movies['primary_genre'] # display primary genres
```

```python
movies.groupby('primary_genre')['Revenue (Millions)'].mean() # group by genre and get the average revenue.
```
