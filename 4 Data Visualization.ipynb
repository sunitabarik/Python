### Bar plots

__Let us plot the count of customers based on the number of lines__

### Pie Charts

__Let us plot the count of customers based on the number of lines__

```python
# makes it convenient to create common layouts of subplots, 
# including the enclosing figure object, in a single call
fig1, ax1 = plt.subplots()


# gives the categories and the number of instances in each category
x = list(df['MultipleLines'].value_counts().index) 
y = list(df['MultipleLines'].value_counts())


#Plots the chart
ax1.pie(y, labels = x, autopct='%1.1f%%', startangle=90)


# ensures that pie is going to be a perfect circle
ax1.axis('equal') 


# tight_layout automatically adjusts subplot params 
# so that the subplot(s) fits in to the figure area. 
# This is an experimental feature and may not work for some cases.
plt.tight_layout()  


# Displays the chart
plt.show() 
```

__Let us plot the count of contract based on the number of lines__

```python
# makes it convenient to create common layouts of subplots, 
# including the enclosing figure object, in a single call
fig1, ax1 = plt.subplots()


# gives the categories and the number of instances in each category
x = list(df['Contract'].value_counts().index) 
y = list(df['Contract'].value_counts())


#Plots the chart
ax1.pie(y, labels = x, autopct='%1.1f%%', startangle=90)


# ensures that pie is going to be a perfect circle
ax1.axis('equal') 


# tight_layout automatically adjusts subplot params 
# so that the subplot(s) fits in to the figure area. 
# This is an experimental feature and may not work for some cases.
plt.tight_layout()  


# Displays the chart
plt.show() 
```

# Matplotlib

This is the primary visualization library of python. However, it is more difficult to use as compared to seaborn

```python
import matplotlib.pyplot as plt

apple_stock_p = pd.read_csv('AAPL.csv')
apple_stock_p.head()
```

```python
plt.plot(apple_stock_p['Adj Close'])
```

```python
# Monthly average
```

```python
apple_stock_p['month'] = [i[1] for i in 
                          apple_stock_p['Date'].str.split('-')]


apple_stock_p['year'] = [i[0] for i in 
                          apple_stock_p['Date'].str.split('-')]
```

```python
apple_stock_p[apple_stock_p['year'] == '2015'].groupby('month').mean()
```

```python
fig = plt.figure()

plt.plot(apple_stock_p[apple_stock_p['year'] 
                       == '2015'].groupby('month').mean()['Adj Close'])

fig.suptitle('APPL Stock Prices Monthly Average 2015', fontsize=20)

plt.xlabel('Month', fontsize=18)

plt.ylabel('Price', fontsize=16)
```

######################################################################

```python
plt.hist(apple_stock_p['Volume'], bins=8)

plt.xlabel('Number of Shares Exchanged', fontsize=18)

plt.ylabel('Frequency', fontsize=16)

plt.show()
```
